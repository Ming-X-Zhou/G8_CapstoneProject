```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.decomposition import IncrementalPCA
import numpy as np
import os
import random
import zipfile
```

```python
df=pd.read_csv('/content/drive/MyDrive/Dataset/Data/Balanced_mineral_data_5000.csv')
df = df.fillna(0)
df=df.drop(['x','y','yinv'], axis=1)
df.head()
```

```python
Y=df['Mineral Phase']
X=df.drop(['Mineral Phase'], axis=1)
df=[]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y, test_size = 0.2, random_state=12)
X_train.shape, X_test.shape
```

```python
n_batches = 100
inc_pca = IncrementalPCA(n_components=154)
for X_batch in np.array_split(X_train, n_batches):
  inc_pca.partial_fit(X_batch)
X_train_pca = inc_pca.transform(X_train)
X_test_pca = inc_pca.transform(X_test)
```

```python
X_train_pca = inc_pca.transform(X_train)
X_test_pca = inc_pca.transform(X_test)
```

```python
inc_pca.explained_variance_ratio_.sum()
```

Incremental PCA is holding 99.99% of varience

```python
svm_classifier = SVC()
svm_classifier.fit(X_train_pca, Y_train)
```

```python
predictions01 = svm_classifier.predict(X_test_pca)

accuracy = accuracy_score(Y_test, predictions01)
print("Accuracy:", accuracy)
```

```python
report = classification_report(Y_test, predictions01)
print(report)
```

```python
svm_classifier_100 = SVC(C=100)
svm_classifier_100.fit(X_train_pca, Y_train)
predictions02 = svm_classifier_100.predict(X_test_pca)

accuracy = accuracy_score(Y_test, predictions02)
print("Accuracy:", accuracy)
```

```python
report = classification_report(Y_test, predictions02)
print(report)
```

```python
svm_classifier_1000 = SVC(C=1000)
svm_classifier_1000.fit(X_train_pca, Y_train)
predictions = svm_classifier_1000.predict(X_test_pca)

accuracy = accuracy_score(Y_test, predictions)
print("Accuracy:", accuracy)
```

```python
report = classification_report(Y_test, predictions)
print(report)
```

```python
# Path to the zip file
zip_file_path = '/content/drive/MyDrive/dataset/CH Counts Data.zip'

# Directory to extract the random files
extraction_directory = 'extracted_files'
os.makedirs(extraction_directory, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Get a list of all file names in the zip
    all_files = zip_ref.namelist()
    random_file = random.choice(all_files)
    zip_ref.extract(random_file, extraction_directory)

# Read the csv file
df = pd.read_csv(os.path.join(extraction_directory, random_file))
df.head()
```
