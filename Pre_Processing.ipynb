```python
import pandas as pd
import random

master_file_path = "/content/drive/MyDrive/random_master_df.csv"

random_master_df = pd.read_csv(master_file_path)


# Initialize a list to store the names of minerals that have already been sampled

output_file_path = "/content/drive/MyDrive/random_master_df2.csv"

count = 2550

# Shuffle the list of training file names (commented out as already shuffled)
#random.shuffle(train_file_names)

# Loop through the list of shuffled training file names
for name in train_file_names[1900:]:
    count += 1
    print(count)

    # Read the current CSV file
    df = pd.read_csv(name)
    df = df.fillna(0)
    df["Mineral Phase"] = df["Mineral Phase"].astype(str)
    n_col = len(df.columns)
    df = df.drop(df.iloc[:, 2800:(n_col-2)], axis=1)
    df = df.drop(df.iloc[:, :126], axis=1)

    # Loop through the list of minerals
    for mineral in mineral_list:
        # Check if the current mineral has already been sampled
        if mineral not in sampled_minerals:
            # Filter the DataFrame for the current mineral
            df_min = df[df["Mineral Phase"].str.lower().str.strip() == mineral.lower().strip()]

            # Check if the number of rows for the current mineral in the master DataFrame is less than 2000
            if len(random_master_df[random_master_df["Mineral Phase"] == mineral]) < 2000:
                # Calculate the number of rows to add, limiting to 100 for each mineral from each file
                rows_to_add = min(100, 2000 - len(random_master_df[random_master_df["Mineral Phase"] == mineral]), len(df_min))

                # Randomly sample rows and concatenate them to the master DataFrame
                sampled_rows = df_min.sample(n=rows_to_add, replace=False)
                random_master_df = pd.concat([random_master_df, sampled_rows], ignore_index=True)

                # Check if 1000 samples have been collected for the current mineral
                if len(random_master_df[random_master_df["Mineral Phase"] == mineral]) >= 2000:
                    # Add the current mineral to the list of sampled minerals
                    sampled_minerals.append(mineral)

    if count % 50 == 0:
        random_master_df.to_csv(output_file_path, index=False)
        print(f"random_master_df saved to {output_file_path} at iteration {count}")
        print(random_master_df["Mineral Phase"].value_counts())

# Print the count of each mineral in the master DataFrame
print(random_master_df["Mineral Phase"].value_counts())

```
