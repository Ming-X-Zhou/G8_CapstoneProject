```python
# Function to Plot training and validation loss/accuracy and learning rate
def history_plot(hist):
    plt.figure(figsize=(15, 5))
    epochs = len(hist.history['loss'])
    # Plot training and validation loss
    plt.subplot(131)
    plt.plot(hist.history['loss'], 'r-', label='Training Loss')
    plt.plot(hist.history['val_loss'], 'b-', label='Validation Loss')
    plt.xlim(0, epochs)
    plt.ylim(0, 2.2)
    plt.xlabel('Epoch')
    plt.ylabel('Loss')
    plt.title('Training and Validation Loss')
    plt.grid(True)
    plt.legend()

    # Plot training and validation accuracy
    plt.subplot(132)
    plt.plot(hist.history['accuracy'], 'r-', label='Training Accuracy')
    plt.plot(hist.history['val_accuracy'], 'b-', label='Validation Accuracy')
    plt.xlim(0, epochs)
    plt.ylim(0, 1.2)
    plt.xlabel('Epoch')
    plt.ylabel('Accuracy')
    plt.title('Training and Validation Accuracy')
    plt.grid(True)
    plt.legend()

    # Plot learning rate
    plt.subplot(133)
    plt.plot(hist.history['lr'], 'g-', label='Learning Rate')
    plt.xlabel('Epoch')
    plt.ylabel('Learning Rate')
    plt.title('Learning Rate Schedule')
    plt.grid(True)
    plt.legend()

    plt.tight_layout()
    plt.show()
```

```python
#Split the dataset into training,validation and test sets
X = minerals_train.iloc[:, :-2]
y = minerals_train['Mineral Phase']

X_train_full, X_test, y_train_full, y_test = train_test_split(X, y, test_size=0.15, random_state=42)

# Split the training data between training and validation set
X_train, X_val, y_train, y_val = train_test_split(X_train_full, y_train_full, test_size=0.15, random_state=42)

# Standarize the features
X_train_scaled = scaler.fit_transform(X_train)
X_val_scaled = scaler.transform(X_val)
X_test_scaled = scaler.transform(X_test)

# Reshape the input data to fit the desired input shape
X_train = X_train.to_numpy().reshape(-1, 4096, 1)
X_val = X_val.to_numpy().reshape(-1, 4096, 1)
X_test = X_test.to_numpy().reshape(-1, 4096, 1)



# Display the shape of the data subsets
print("Training data shape:", X_train.shape)
print("Validation data shape:", X_val.shape)
print("Testing data shape:", X_test.shape)


#onehot_encoder = OneHotEncoder(sparse_output=False)

# Convert pandas Series to NumPy arrays
#y_train = y_train.to_numpy().reshape(-1, 1)
#y_val = y_val.to_numpy().reshape(-1, 1)
#y_test = y_test.to_numpy().reshape(-1, 1)

# Perform one-hot encoding
#y_train_encoded = onehot_encoder.fit_transform(y_train)
#y_val_encoded = onehot_encoder.transform(y_val)
#y_test_encoded = onehot_encoder.transform(y_test)
```

```python
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, BatchNormalization, Activation

# Define a function to create the CNN model
def create_cnn(num_layers=3, kernel_size=2, num_kernel=64, activation_function='relu', num_minerals=38,input_shape=(4096, 1)):
    model = Sequential()

    # First Convolutional Layer
    model.add(Conv1D(num_kernel, kernel_size=kernel_size, input_shape=input_shape, padding='same'))
    model.add(BatchNormalization())
    model.add(Activation(activation_function))
    model.add(MaxPooling1D(pool_size=2))
    model.add(layers.Dropout(0.4))

    # Second Convolutional Layer
    model.add(Conv1D(32, kernel_size=kernel_size, padding='same'))
    model.add(BatchNormalization())
    model.add(Activation(activation_function))
    model.add(MaxPooling1D(pool_size=2))
    model.add(layers.Dropout(0.3))

    if num_layers == 3:
        # Third Convolutional Layer
        model.add(Conv1D(32, kernel_size=kernel_size, padding='same'))
        model.add(BatchNormalization())
        model.add(Activation(activation_function))
        model.add(MaxPooling1D(pool_size=2))
        model.add(layers.Dropout(0.2))

    # Fully Connected Layer
    model.add(Flatten())

    #Dense Layer 1
    model.add(Dense(64,kernel_initializer='glorot_normal'))
    model.add(BatchNormalization())
    model.add(Activation(activation_function))
    model.add(Dropout(0.4))

    # Output Layer
    model.add(Dense(num_minerals, activation='softmax',kernel_initializer='glorot_normal'))

    # Compile the model
    model.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=0.001), loss='categorical_crossentropy', metrics=['accuracy'])

    return model
```

```python
import time

# Build the model with the best hyperparameters
# Best Parameters:{'activation_function': 'relu', 'num_layers': 3, 'kernel_size': (3, 3), 'num_kernel': 64}
cnn_model = create_cnn()

# Model Summary
cnn_model.summary()

# Define early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=10, restore_best_weights=True)

# Define performance-based learning rate scheduling
lr_schhedule = tf.keras.callbacks.ReduceLROnPlateau(monitor='val_loss', factor=0.2, patience=5)

# Start the timer
start_time_cnn = time.time()

# Fit the training data
history_cnn = cnn_model.fit(X_train, y_train_encoded, validation_data=(X_val, y_val_encoded), batch_size=32, epochs=100,
                          callbacks=[early_stopping,ModelCheckpoint("cnn", save_best_only=True),lr_schhedule])

# Stop the timer and calculate the model training time
end_time_cnn = time.time()
training_time_cnn = (end_time_cnn - start_time_cnn) / 60

# Save the model in TensorFlow save model format
cnn_model.save("/content/drive/MyDrive/caps_cnn_model_test", save_format='tf')
```

```python
# Obtain the prediction
cnn_train_predict = cnn_model.predict(X_train)
cnn_test_predict = cnn_model.predict(X_test)

# Calculate the accuracy on training and test set
cnn_train_accuracy = accuracy_score(np.argmax(y_train_encoded, axis=1), np.argmax(cnn_train_predict, axis=1))
cnn_test_accuracy = accuracy_score(np.argmax(y_test_encoded, axis=1), np.argmax(cnn_test_predict, axis=1))


print(f"Computation time for a CNN model is {training_time_cnn:.2f} minutes\n\n\n")
```
