Extracting 50 files in random from zip folder for testing

```python
# Path to the zip file
zip_file_path = '/Volumes/Backup Plus/mac/CH Counts Data.zip/CH Counts Data.zip'

# Directory to extract the random files
extraction_directory = 'extracted_files'
os.makedirs(extraction_directory, exist_ok=True)

# Extract the zip file
with zipfile.ZipFile(zip_file_path, 'r') as zip_ref:
    # Get a list of all file names in the zip
    all_files = zip_ref.namelist()

    # Choose 50 random files (or all files if there are fewer than 50)
    selected_files = random.sample(all_files, min(50, len(all_files)))

    # Extract the selected files
    for file_name in selected_files:
        zip_ref.extract(file_name, extraction_directory)

print("Selected files extracted to:", extraction_directory)
```

Specifying the classification to be used for testing

```python
def Class_select(df):
  classifiers_to_keep = {'Fe-Oxide', 'AlFe-Silicate', 'Barite', 'Fe-Silicate', 'AlMn-Silicate', 'Ilmenite', 'Amphibole', 'Monazite', 'Quartz', 'Xenotime'}
  df = df[df['Mineral Phase'].isin(classifiers_to_keep)]
  return df
```

Droping features that are do not contain much information for classification, then the dataframe is sent to class_select function to extract rows  we need for testing. The final dataframe is concatenated.

```python
def col_drop():
  dfs=[]
  for f in files:
    path='/Users/bhargavakolli/Documents/Sem_3/Capstone/EDA/extracted_files/CH Counts Data/'+ f
    df= pd.read_csv(path)
    #print(df.columns[2501:-1])
    df.drop(df.columns[2501:-1], axis=1, inplace=True)
    df.drop(df.columns[:123],axis=1,inplace=True)
    print(path)
    new_df=Class_select(df)
    print(new_df)
    dfs.append(new_df)
  df_concat= pd.concat(dfs, ignore_index=True)
  concat_df_path='/Users/bhargavakolli/Documents/Sem_3/Capstone/EDA/concat50.csv'
  df_concat.to_csv(concat_df_path, index=False)
col_drop()
```
